```python
for i in df1.Marque.unique():
    bd_Marque_i=df1.loc[df1.Marque==i,['Marque','Modèle','Prix']]
    q1 = bd_Marque_i['Prix'].quantile(0.25)
    q3 = bd_Marque_i['Prix'].quantile(0.75)
    iqr = q3-q1 #Interquartile range
    fence_low  = q1-1.5*iqr
    fence_high = q3+1.5*iqr
    for j in bd_Marque_i.index:
        if((bd_Marque_i.loc[j,'Prix']<fence_low) or (bd_Marque_i.loc[j,'Prix']>fence_high)):
            indice=j
            df1.drop(indice, inplace=True)
```
