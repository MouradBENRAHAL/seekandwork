# Data Cleaning

# Année

```python
for i in base.index:
    if (base.loc[i,'Année']<0):
        base.loc[i,'Année']=abs(base.loc[i,'Année'])
```

```python
import time 
this_year=time.localtime().tm_year
for i in base.index:
    if (base.loc[i,'Année']<1900) or (base.loc[i,'Année']>this_year):
        base.drop(i, inplace=True)
```

# Kilométrage

```python
for i in base.index:
    if(base.loc[i,'Kilométrage']<1000):
        base.loc[i,'Kilométrage']=base.loc[i,'Kilométrage']*1000
```

```python
for i in base.index:
    if (base.loc[i,'Kilométrage']<0):
        base.loc[i,'Kilométrage']=abs(base.loc[i,'Kilométrage'])
```

```python
for i in base.index:
    if(base.loc[i,'Kilométrage']>1000000):
        base.loc[i,'Kilométrage']=base.loc[i,'Kilométrage']/1000
```

```python
for i in base.index:
    if(base.loc[i,'Kilométrage']>500000) or (base.loc[i,'Kilométrage']<5000) :
        base.drop(i, inplace=True)
```

```python
a=base[['Marque','Kilométrage']]
plt.figure(figsize=(50, 10))
sns.boxplot(x='Marque', y='Kilométrage', data=a)
plt.show()
```

```python
for i in base.Marque.unique():
    bd_Marque_i=base.loc[base.Marque==i,['Marque','Kilométrage']]
    q1 = bd_Marque_i['Kilométrage'].quantile(0.25)
    q3 = bd_Marque_i['Kilométrage'].quantile(0.75)
    iqr = q3-q1 #Interquartile range
    fence_low  = q1-1.5*iqr
    fence_high = q3+1.5*iqr
    for j in bd_Marque_i.index:
        if((bd_Marque_i.loc[j,'Kilométrage']<fence_low) or (bd_Marque_i.loc[j,'Kilométrage']>fence_high)):
            indice=j
            base.drop(indice, inplace=True)
```

```python
a=base[['Marque','Kilométrage']]
plt.figure(figsize=(50, 10))
sns.boxplot(x='Marque', y='Kilométrage', data=a)
plt.show()
```

```python
base['Kilométrage'].describe()
```

# Prix

```python
for i in base.index:
    base.loc[i,'Prix']=str(base['Prix'][i]).replace('\xa0','')
```

```python
for i in base.index:
    base.loc[i,'Prix']=str(base['Prix'][i]).replace('\u202f','')
```

```python
for i in base.index:
    if (base.loc[i,'Prix']=="nan"):
        base.drop(i, inplace=True)   
```

```python
base[["Prix"]] = base[["Prix"]].apply(pd.to_numeric)
```

```python
for i in base.index:
    if(base.loc[i,'Prix']<1000):
        base.loc[i,'Prix']=base.loc[i,'Prix']*1000
```

```python
for i in base.index:
    if(base.loc[i,'Prix']>1000000):
        base.loc[i,'Prix']=base.loc[i,'Prix']/1000
```

```python
for i in base.index:
    if(base.loc[i,'Prix']==0):
        base.drop(i, inplace=True) 
```

```python
a=base[['Marque','Prix']]
plt.figure(figsize=(50, 10))
sns.boxplot(x='Marque', y='Prix', data=a)
plt.show()
```

```python
for i in base.Marque.unique():
    bd_Marque_i=base.loc[base.Marque==i,['Marque','Prix']]
    q1 = bd_Marque_i['Prix'].quantile(0.25)
    q3 = bd_Marque_i['Prix'].quantile(0.75)
    iqr = q3-q1 #Interquartile range
    fence_low  = q1-1.5*iqr
    fence_high = q3+1.5*iqr
    for j in bd_Marque_i.index:
        if((bd_Marque_i.loc[j,'Prix']<fence_low) or (bd_Marque_i.loc[j,'Prix']>fence_high)):
            indice=j
            base.drop(indice, inplace=True)
```

```python
a=base[['Marque','Prix']]
plt.figure(figsize=(50, 10))
sns.boxplot(x='Marque', y='Prix', data=a)
plt.show()
```

```python
a=base[['Modèle','Prix']]
plt.figure(figsize=(50, 10))
sns.boxplot(x='Modèle', y='Prix', data=a)
plt.show()
```
