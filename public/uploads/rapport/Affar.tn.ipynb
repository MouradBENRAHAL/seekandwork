# Scrapper tous les liens


```python
link='https://www.affare.tn/petites-annonces/tunisie/voiture-neuve-occassion-prix-tayara-a-vendre'
i=0
links=[]
while True:
    print(i)
    url=link+"?o="+str(i)
    page= requests.get(url)
    soup = BeautifulSoup(page.text, 'html.parser')
    l=soup.findAll('a',class_="saz")
    if len(l)==0:
        break
    else:
        for j in l:
            lien="https://www.affare.tn/"+str(j.get('href'))
            links.append(lien)
    i+=1 
```

```python
def scrap (a):
    page= requests.get(a)
    soup = BeautifulSoup(page.text, 'html.parser')
    try:
        prix=soup.find('span',class_="price2").text
        info=soup.findAll(class_='box-params-inner\n')
        for k in info:
            print(k.text,"\n")
    except:
        pass
    return prix
```

```python
for i in links:
    try:
        S=scrap(i) 
        print(S)
    except :
        pass
```

```python
df=pd.DataFrame(columns=['Energie','Année','Kilométrage','Puissance','Mise en circulation','Boite','Prix'])
i=0
for link in links:
    print(i,'/',len(links))
    df.loc[i,'lien']=link
    page= requests.get(link)
    soup = BeautifulSoup(page.text, 'html.parser')
    info1=soup.find_all(class_='box-params-inner')
    try :
        df.loc[i,'Energie']=info1[1].text
    except :
        print('error in',i)
        pass
    try :
        df.loc[i,'Année']=info1[2].text
    except :
        print('error in',i)
        pass
    try :    
        df.loc[i,'Kilométrage']=info1[3].text
    except :
        print('error in',i)
        pass
    try :
        df.loc[i,'Puissance']=info1[4].text
    except :
        print('error in',i)
        pass
    try :
        df.loc[i,'Mise en circulation']=info1[5].text
    except :
        print('error in',i)
        pass
    try :
        df.loc[i,'Boite']=info1[6].text
    except :
        print('error in',i)
        pass
    try :
        df.loc[i,'Prix']=soup.find('span',class_="price2").text   
    except :
        print('error in',i)
        pass
    i+=1
    if i>20:
        break
```
