```python
#remplir la colonne model ajoutée au df
for i in df.index:
    print(i)
    mo=""
    rat=0
    for model in sorted(marque_modele.loc[marque_modele.Marque==df.marque[i],'Modèle'],key=len):
        if fuzz.partial_ratio(model.lower().replace(' ','').replace('é','e'),df['Modèle'][i].lower().replace(' ','').replace('é','e'))>=rat:
            mo=model
            rat=fuzz.partial_ratio(model.lower(),df.loc[i,'Modèle'].lower())
    df.loc[i,'model']=mo
```

```python
from sklearn import linear_model
from sklearn import linear_model
def prediction(css_data):
    mar=list(css_data.marque.unique())
    print(mar)
    ch_marque=input("choisir la marque de votre voiture de cette liste ")
    dic={}
    for ma in css_data.marque.unique() :
        dic[ma]=list(css_data.loc[css_data.marque==ma,'model'].unique())
    while(ch_marque in mar):
        print(list(dic[ch_marque]))
        ch_modele=input("entrez le modèle de votre voiture de cette liste")
        while(ch_modele in list(dic[ch_marque])):
            specified_data=css_data.loc[(css_data.marque==ch_marque)&(css_data.model==ch_modele)]
            dummy=pd.get_dummies(specified_data["Energie"])
            specified_data=pd.concat([specified_data,dummy],axis=1)
            reg=linear_model.LinearRegression()
            a=int(input("entrez l'année de mise en circulation de la voiture"))
            p=int(input("entrez la puissance fiscale de la voiture"))
            k=int(input("entrez le kilometrage de la voiture"))
            carbu=input("entrez le carburant de la voiture essence/diesel ")
            if(carbu=="essence"):
                e=1
                reg.fit(specified_data[['Année','Puissance','Kilométrage','Essence']],specified_data.Prix)
                prix_estimé=reg.predict([[a,p,k,e]]) 
            else:
                d=1
                reg.fit(specified_data[['Année','Puissance','Kilométrage','Diesel']],specified_data.Prix)
                prix_estimé=reg.predict([[a,p,k,d]])   
            return prix_estimé
            
        
```

```python
df2=df2[['Kilométrage','Année','marque','model','Energie','Prix']]
def pred_Gradient(km,an,ma,mo,ca):
    cat_vars=['marque','model','Energie']
    data=df2.loc[df2.model==mo]
    for var in cat_vars:
        cat_list='var'+'_'+var
        cat_list = pd.get_dummies(data[var], prefix=var)
        data1=data.join(cat_list)
        data=data1
    ####### Data & Target  
    col=data.columns
    col=col.drop(['model','marque','Energie','Prix'])
    X=data[col]
    Y=data['Prix']
    ####### Entrer les données
    XX=pd.DataFrame(columns=X.columns)
    XX.loc[0]=np.zeros(XX.shape[1])
    XX.loc[0,'Kilométrage']=int(km)
    XX.loc[0,'Année']=int(an)
    XX.loc[0,'marque_'+ma]=1
    XX.loc[0,'model_'+mo]=1
    XX.loc[0,'Energie_'+ca]=1
    params = {'n_estimators': 500, 'max_depth': 4, 'min_samples_split': 2,'learning_rate': 0.01, 'loss': 'ls'}
    clf = ensemble.GradientBoostingRegressor(**params)
    clf.fit(X, Y)    
    return clf.predict(XX)
```

```python
import sklearn.metrics as metrics
def regression_results(y_true, y_pred):

    # Regression metrics
    explained_variance=metrics.explained_variance_score(y_true, y_pred)
    mean_absolute_error=metrics.mean_absolute_error(y_true, y_pred) 
    mse=metrics.mean_squared_error(y_true, y_pred) 
    try:
        mean_squared_log_error=metrics.mean_squared_log_error(y_true, y_pred)
    except:
        mean_squared_log_error=0
        pass    
    median_absolute_error=metrics.median_absolute_error(y_true, y_pred)
    r2=metrics.r2_score(y_true, y_pred)

    print('explained_variance: ', round(explained_variance,4))    
    print('mean_squared_log_error: ', round(mean_squared_log_error,4))
    print('r2: ', round(r2,4))
    print('MAE: ', round(mean_absolute_error,4))
    print('MSE: ', round(mse,4))
    print('RMSE: ', round(np.sqrt(mse),4))
```
